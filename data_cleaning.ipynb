```python
def convert_year(year):
    if len(year) > 4:
        year = year.split('-')
        return  int(str(year[0])[:2]  + str(year[1]))
    return year
```

```python
def keep_most_recent_date(df):
    df['DataSourceYears'] = df['DataSourceYears'].apply(convert_year).apply(int)
    df = df.sort_values(by=['CountryName', 'DataSourceYears'], ascending=[True, False])
    df = df.drop_duplicates(subset='CountryName', keep='first')
    return df
```

```python
wasting = keep_most_recent_date(wasting)
stunting = keep_most_recent_date(stunting)
overweight = keep_most_recent_date(overweight)

```

```python
overweight.to_csv('cleaned_data/overweight_cleaned.csv')
stunting.to_csv('cleaned_data/stunting_cleaned.csv')
wasting.to_csv('cleaned_data/wasting_cleaned.csv')
```
